## Feature creation 
#### Physics-based features were created before the train-test split, as they rely on domain equations and do not cause data leakage.

```python
import pandas as pd
import numpy as np

# ─── STEP 1: INITIALIZE COPIES OF THE DELTA COMPONENTS ───
# We group by Field_Unique_ID so that the math doesn't bleed between different fields
grouped = df.groupby('Field_Unique_ID')

delta_P  = grouped['Current_Reservoir_Pressure_psi'].diff()
delta_R  = grouped['Producing_GOR_R_scf_stb'].diff()
delta_Np = grouped['Cumulative_Oil_Np_stb'].diff()

# Small safety buffer to avoid dividing by absolute zero on initial timesteps
epsilon = 1e-6

# ─── STEP 2: CALCULATE THE FEATURE EQUATIONS ───

# 1. Dimensionless Depletion Ratio (P / Pi)
df['P_Pi_ratio'] = df['Current_Reservoir_Pressure_psi'] / df['Initial_Reservoir_Pressure_psi']

# 2. Instantaneous Water Cut (fw)
df['Water_Cut'] = df['Water_Production_Rate_qw_bbl_day'] / (
    df['Oil_Production_Rate_qo_bbl_day'] + df['Water_Production_Rate_qw_bbl_day'] + epsilon
)

# 3. Diagnostic Derivative: Pressure Decline per unit of Oil Produced (dP / dNp)
df['dP_dNp'] = delta_P / (delta_Np + epsilon)

# 4. Diagnostic Derivative: GOR Change per unit of Oil Produced (dR / dNp)
df['dR_dNp'] = delta_R / (delta_Np + epsilon)

#5.
df["Below_Bubble_Point"] = (
    df["Current_Reservoir_Pressure_psi"] < df["Bubble_Point_Pressure_psi"]
).astype(int)

#6.
df["GOR_by_Rsi"] = (
    df["Producing_GOR_R_scf_stb"] / df["Initial_Solution_GOR_scf_stb"]
)


# ─── STEP 3: CLEAN TIMESTEP BOUNDARY NOISE ───
# The very first month of production has no "previous month," resulting in a NaN value for the deltas.
# We backfill (bfill) Month 1 with Month 2's trend data within each field to keep the matrix clean.
df['dP_dNp'] = df.groupby('Field_Unique_ID')['dP_dNp'].bfill()
df['dR_dNp'] = df.groupby('Field_Unique_ID')['dR_dNp'].bfill()

# ─── STEP 4: VERIFICATION INSPECTION ───
print("Feature Engineering Complete. Checking for missing values:")
print(df[['P_Pi_ratio', 'Water_Cut', 'dP_dNp', 'dR_dNp']].isna().sum())
```

## Train Test Spiliting
### Two field IDs from randomly selected drive mechanisms were separated and used as the test set to evaluate the model’s performance.

```python
import numpy as np
import pandas as pd

# Set seed for reproducibility so your randomized splits stay consistent across runs
np.random.seed(42)


# Lists to gather our dynamically chosen train and test row indices
train_indices = []
test_indices = []

# 2. Loop through each drive mechanism and pick 3 random fields
unique_drives = df['Drive_Mechanism'].unique()

for drive in unique_drives:
    # Isolate all unique field IDs for this specific drive mechanism
    drive_fields = df[df['Drive_Mechanism'] == drive]['Field_Unique_ID'].unique()
    
    # RANDOMLY select exactly 3 field IDs out of the 13 available
    test_fields_chosen = np.random.choice(drive_fields, size=3, replace=False)
    
    # Separate rows belonging to the chosen test fields vs training fields
    drive_rows = df[df['Drive_Mechanism'] == drive]
    
    test_rows = drive_rows[drive_rows['Field_Unique_ID'].isin(test_fields_chosen)]
    train_rows = drive_rows[~drive_rows['Field_Unique_ID'].isin(test_fields_chosen)]
    
    # Save the index locations
    test_indices.extend(test_rows.index)
    train_indices.extend(train_rows.index)

# 3. Create the clean randomized Train and Test DataFrames using the gathered indices
train_df = df.loc[train_indices].copy()
test_df  = df.loc[test_indices].copy()

print(train_df.shape)
print(test_df.shape)
```

## Separation into X and y and feature selecion
### Separate input features (X) and target variable (y), and select the relevant features for model training. 

```python
# 3. Separate Features (X) and Target (y)
features = [
    'Water_Cut',
    'P_Pi_ratio',
    'GOR_by_Rsi',
    'Below_Bubble_Point',
    'dP_dNp',
    'dR_dNp',
    ]

X_train = train_df[features]
y_train = train_df['Drive_Label']

X_test  = test_df[features]
y_test  = test_df['Drive_Label']
```

# Model Training
### ML Algorithm - Random forest Classifier

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
 

# 1. Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# 2. Train (Fit) the Model on your Training Set
rf_model.fit(X_train, y_train)
print("Model Training Complete!")
```

# Model Evaluation

```python
# Predict on the Test Set
y_pred = rf_model.predict(X_test)          # Gives the rigid 0, 1, 2, 3 labels
y_proba = rf_model.predict_proba(X_test)    # Gives the combination percentage matrix

y_pred_df = pd.DataFrame(y_pred, columns=['Predicted_Label'])
y_proba_df = pd.DataFrame(y_proba, columns=[
    'Expansion_Prob', 
    'Depletion_Prob', 
    'GasCap_Prob', 
    'WaterDrive_Prob'
])

print(y_pred_df.head())
print(y_proba_df.head(14))
```

```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# 1. Initialize lists to hold true and predicted labels at the ASSET FIELD level
y_field_true = []
y_field_pred = []

# Get the list of the 12 unique unseen fields in your test dataframe
unique_test_fields = test_df['Field_Unique_ID'].unique()

# 2. Iterate through each unique field history as a single object
for field_id in unique_test_fields:
    # Isolate all 200 rows belonging to this specific field
    field_rows = test_df[test_df['Field_Unique_ID'] == field_id]
    
    # Extract features for this field
    X_field = field_rows[features]
    
    # Capture the actual true label of the field (constant across its 200 months)
    actual_field_label = field_rows['Drive_Label'].iloc[0]
    y_field_true.append(actual_field_label)
    
    # Predict the row-by-row probabilities for this specific field timeline
    row_probabilities = rf_model.predict_proba(X_field)
    
    # PHYSICS COMPRESSION: Average the 200 months of probabilities into 1 vector
    average_field_probabilities = np.mean(row_probabilities, axis=0)
    
    # The final field prediction is the index with the highest average probability
    final_field_prediction = np.argmax(average_field_probabilities)
    y_field_pred.append(final_field_prediction)

# Convert evaluation lists to numpy arrays for scikit-learn
y_field_true = np.array(y_field_true)
y_field_pred = np.array(y_field_pred)

# 3. Generate the Asset-Level Classification Report DataFrame
field_report = classification_report(
    y_field_true,
    y_field_pred,
    labels=[0, 1, 2, 3],
    target_names=['Expansion (0)', 'Depletion (1)', 'Gas Cap (2)', 'Water Drive (3)'],
    zero_division=0,
    output_dict=True
)

field_report_df = pd.DataFrame(field_report).transpose()

print("=" * 65)
print("             FIELD-LEVEL CLASSIFICATION REPORT             ")
print("=" * 65)
print(field_report_df)
print("-" * 65)

```

```python
# 4. Generate and Plot the Asset-Level Confusion Matrix
from sklearn.metrics import confusion_matrix
class_names = ['Expansion', 'Depletion', 'Gas Cap', 'Water']
cm_fields = confusion_matrix(y_field_true, y_field_pred, labels=[0, 1, 2, 3])

plt.figure(figsize=(8, 6))
sns.heatmap(cm_fields, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)

plt.title('Asset-Level Drive Mechanism Classifier - Confusion Matrix', fontsize=12, fontweight='bold')
plt.ylabel('True Field Mechanism')
plt.xlabel('Predicted Field Diagnosis')
plt.tight_layout()
plt.show()
```
